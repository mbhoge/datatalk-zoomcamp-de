```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
engine.connect()
```

# Trip Data

```python
pd.read_sql('SELECT index, "VendorID", lpep_pickup_datetime, lpep_dropoff_datetime, store_and_fwd_flag, "RatecodeID", "PULocationID", "DOLocationID", passenger_count, trip_distance, fare_amount, extra, mta_tax, tip_amount, tolls_amount, ehail_fee, improvement_surcharge, total_amount, payment_type, trip_type, congestion_surcharge FROM public.green_tripdata',con=engine)
```

# Question 3: Count records  (Multiple choice)

# How many taxi trips were totally made on January 15?


```python
pd.read_sql("SELECT count(1) FROM public.green_tripdata where substring(lpep_pickup_datetime,6,5) = '01-15'",con=engine)
```

# Question 4: Largest trip for each day (Multiple choice)

# Which was the day with the largest trip distance?

```python
pd.read_sql("select to_timestamp(lpep_dropoff_datetime,'YYYY-MM-DD') as trip_day, max(trip_distance) as highest_distance FROM public.green_tripdata group by trip_day order by highest_distance desc fetch first 1 rows only",con=engine)
```

# Question 5: The number of passengers  (Multiple choice)

# In 2019-01-01 how many trips had 2 and 3 passengers?

```python
pd.read_sql("select passenger_count, count(1) from public.green_tripdata where to_timestamp(lpep_pickup_datetime,'YYYY-MM-DD')='2019-01-01' group by passenger_count having passenger_count=2 or passenger_count=3 ",con=engine)
```

# Question 6: Largest tip (Multiple choice)

# For the passengers picked up in the Astoria Zone which was the drop up zone that had the largest tip?

```python
sql="""SELECT z."Zone" from (SELECT Y."PULocationID", max(tip_amount) as high_tip, "DOLocationID" FROM public.zones X inner join public.green_tripdata Y on X."LocationID" = Y."PULocationID" group by Y."PULocationID","DOLocationID",X."Zone" having X."Zone" = 'Astoria' order by high_tip desc fetch first 1 rows only) high_tip_loc inner join public.zones z on high_tip_loc."DOLocationID" = z."LocationID";"""
```

```python
pd.read_sql(sql,con=engine)
```
